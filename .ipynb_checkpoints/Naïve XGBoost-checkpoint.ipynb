def H_builder(sample_indices1, sample_indices2, yd):
    H_n1n2 = 0
    ind_n1 = sample_indices1
    ind_n2 = sample_indices2
    ind = []
    for i1, i2 in ind_n1:
        for i11, i22 in ind_n2:
            if i1==i11:
                ind.append([i1, i11, i2, i22])
                H_n1n2 += np.dot(yd[i2], yd[i22])
    return H_n1n2, ind

#@njit
def my_H_builder(sample_indices1, sample_indices2, yd):
    H_n1n2 = 0
    n1 = len(sample_indices1)
    n2 = len(sample_indices2)
    
    si1 = sample_indices1[np.argsort(sample_indices1[:,0])]
    si2 = sample_indices2[np.argsort(sample_indices2[:,0])]
    #print(si1, si2)
    index_1=0
    index_2=0
    ind = []
    while(index_1<n1 and index_2<n2):
        if si1[index_1][0] == si2[index_2][0] :
            ind.append([si1[index_1][0], si2[index_2][0],
                        si1[index_1][1],
                        si2[index_2][1]])
            index_2+=1
        elif si1[index_1][0] > si2[index_2][0] :
            index_2+=1
            #index_1 =0
        elif si1[index_1][0] < si2[index_2][0]:
            index_1+=1
            index_2=0
            
    for _,_, i2, i22 in ind:###parcours sur les triplets
        #print(np.dot(yd[i2], yd[i22]))
        H_n1n2 += np.dot(yd[i2], yd[i22])
        
    return H_n1n2, ind
