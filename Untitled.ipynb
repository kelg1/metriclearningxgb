### Toy Dataset with 4 features

```python
X = np.random.rand(30,18)
y = []
for i in range(X.shape[0]):
    y.append(X[i][0]**2+X[i][1]*2+np.exp(X[i][2]-X[i][3])+np.linalg.norm(X[i,4:]))
y = np.asarray(y)
y = y.reshape(-1,1)
```

```python
X, Xd, y, yd = train_test_split(X,y)
```

```python
gb = GradientBoostingRegressor(learning_rate=1, max_iter=8,
                 max_leaf_nodes=8, max_depth=2, min_samples_leaf=20,
                 l2_regularization=0.1, max_bins=10, max_no_improvement=50,
                 validation_split=.1, scoring='neg_mean_squared_error',
                 tol=1e-4, verbose=1, random_state=0)
```
